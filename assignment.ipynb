4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C','D']
values = {
    'A': np.random.randint(1, 6, size=100),
    'B': np.random.randint(2, 10, size=100),
    'C': np.random.randint(7,12, size=100),
    'D': np.random.randint(0, 5, size=100)
}

data = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate([values[category] for category in categories])
})
plt.figure(figsize=(10,4))

plt.subplot(1, 2, 1)
sns.boxplot(x='Category', y='Value', data=data)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')

plt.subplot(1, 2, 2) 
sns.violinplot(x='Category', y='Value', data=data)
plt.title('Violin Plot')
plt.xlabel('Category')
plt.ylabel('Value')


```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a

```python

X1 = np.random.normal(0, 1, 100)
X2 = 0.1 * X1 + np.random.normal(0, 0.1, 100) 
X3 = 0.6 * X1 - 0.6 * X2 + np.random.normal(0, 0.1, 100)
X4 = np.random.rand(100) 
data5 = pd.DataFrame({
    'Feature1': X1,
    'Feature2': X2,
    'Feature3': X3,
    'Feature4': X4
})
plt.figure()
sns.heatmap(data5.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(sqrt(x2+y2))

```python
import plotly.graph_objects as go
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
df

fig=go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(
    title='3D Surface Plot of z=sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    coloraxis_colorbar=dict(
        title='Z Value'
    )
)
```

BOKEH ASSIGNMENT:

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, output_notebook, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
output_notebook()
fig = figure(title='Sin Wave', x_axis_label='x', y_axis_label='value',  width=800, height=400)
fig.line(x, y, legend_label='sin wave')
show(fig)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.palettes import Category10

np.random.seed(42)

n = 100
x = np.random.uniform(0, 10, n)
y = np.random.uniform(0, 10, n)
sizes = np.random.uniform(10, 50, n)

possible_colors = ['red', 'black', 'pink', 'orange', 'blue', 'yellow']

colors = np.random.choice(possible_colors, n)

output_notebook()

p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='x', y_axis_label='y',  width=800, height=400)

p.scatter(x, y, size=sizes, color=colors, fill_alpha=0.6)

show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python

from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
xx = xx.flatten()
yy = yy.flatten()
colors = data_heatmap.flatten()

output_notebook()
source = ColumnDataSource(data=dict(x=xx, y=yy, color=colors))
mapper = linear_cmap(field_name='color', palette='Viridis256', low=min(colors), high=max(colors))
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',  width=800, height=400)

p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=mapper, line_color=None)

show(p)
```
